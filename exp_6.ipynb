**Experiment 6:**  Implement Clustering-Based Detection of Abnormal or Abusive Patterns.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Sample user behaviour data
X = np.array([
    [5,10],
    [6,12],
    [7,14],
    [8,15],
    [9,16],
    [50,80],
    [55,85],
    [60,90],
    [58,88],
    [65,95]
])
```

```python
kmeans = KMeans(n_clusters=2)

kmeans.fit(X)

labels = kmeans.labels_

print("Cluster labels:", labels)
```

```python
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)

plt.xlabel("User Activity")
plt.ylabel("Message Frequency")
plt.title("Detection of Abnormal or Abusive Patterns")

plt.show()
```
